```python
def choose_models(nstop, X_trntst, Y_trntst):
    rsq_trn = pd.DataFrame(np.zeros((nstop, 3)), columns =['NN','GB','RF'])
    rsq_tst = pd.DataFrame(np.zeros((nstop, 3)), columns =['NN','GB','RF'])
    rsq_validate = pd.DataFrame(np.zeros((1, 3)), columns =['NN','GB','RF'])
    
    for n in range(nstop):
        X_trn, X_tst, Y_trn, Y_tst = train_test_split(X_trntst, Y_trntst, test_size = .3)
        X_trn_save = X_trn.copy()
        X_tst_save = X_tst.copy()
        Y_trn_save = Y_trn.copy()
        Y_tst_save = Y_tst.copy()
 


    # Neural net   
        NN = MLPClassifier(hidden_layer_sizes=(4,),activation='tanh',solver='adam',
                          learning_rate='adaptive',max_iter=10000,learning_rate_init=.01,alpha=.01)
        NN.fit(X_trn,Y_trn)
        predict_NN_trn = NN.predict(X_trn)
        predict_NN_tst = NN.predict(X_tst)
        rsq_trn.loc[n,'NN'] = f1_score(Y_trn,predict_NN_trn, average='micro')
        rsq_tst.loc[n,'NN'] = f1_score(Y_tst,predict_NN_tst, average='micro')

        X_trn = X_trn_save
        X_tst = X_tst_save
        Y_trn = Y_trn_save
        Y_trn = Y_trn_save


    # Gradient boosting tree    
        params = {'n_estimators':300, 'max_depth':5, 'min_samples_split':20, 'learning_rate':.2}
        GB = GradientBoostingClassifier(**params)
        GB.fit(X_trn,Y_trn)
        predict_GB_trn = GB.predict(X_trn)
        predict_GB_tst = GB.predict(X_tst)
        rsq_trn.loc[n,'GB'] = f1_score(Y_trn,predict_GB_trn, average='micro')
        rsq_tst.loc[n,'GB'] = f1_score(Y_tst,predict_GB_tst, average='micro')

        X_trn = X_trn_save
        X_tst = X_tst_save
        Y_trn = Y_trn_save
        Y_tst = Y_tst_save


    # Random forest
        RF = RandomForestClassifier(n_estimators=10, max_depth=20)    
        RF.fit(X_trn,Y_trn)
        predict_RF_trn = RF.predict(X_trn)
        predict_RF_tst = RF.predict(X_tst)
        rsq_trn.loc[n,'RF'] = f1_score(Y_trn,predict_RF_trn, average='micro')
        rsq_tst.loc[n,'RF'] = f1_score(Y_tst,predict_RF_tst, average='micro')
    
    return rsq_trn,rsq_tst
```
